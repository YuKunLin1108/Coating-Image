```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 讀取圖檔
img = cv2.imread(img_path)

# 轉為灰階圖片
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 計算直方圖每個 bin 的數值
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])

# 畫出直方圖
plt.hist(gray.ravel(), 256, [0, 256])
plt.figure()
plt.plot(range(1,257), hist)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread(img_path)

# 畫出 RGB 三種顏色的分佈圖
color = ('b','g','r')
for i, col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0, 256])
    plt.plot(histr, color = col)
    plt.xlim([0, 256])
plt.show()
```

```python
(b, g, r) = cv2.split(img)
bH = cv2.equalizeHist(b)
gH = cv2.equalizeHist(g)
rH = cv2.equalizeHist(r)
img_resize = cv2.resize(img, (500, 500))
frameH = cv2.merge((bH, gH, rH))
frameH0 = cv2.resize(frameH, (500, 500))
# cv2.imshow('Img', img_resize)
# cv2.imshow('Frame4', frameH0)

img_bH = cv2.resize(bH, (500, 500))
cv2.imshow('img_bH', img_bH)
img_gH = cv2.resize(gH, (500, 500))
cv2.imshow('img_gH', img_gH)
img_rH = cv2.resize(rH, (500, 500))
cv2.imshow('img_rH', img_rH)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def get_edge(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = cv2.resize(gray, (500, 500))
    blur = cv2.GaussianBlur(gray, (13, 13), 0)
    canny = cv2.Canny(blur, 100, 150)
    return canny
#----------------------------#
img = cv2.imread(img_path)
edge = get_edge(img)
cv2.imshow('Edge', edge)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
